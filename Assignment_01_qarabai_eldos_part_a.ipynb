```python
import numpy as np

Resources_Matrix = np.array([[10, 20], [30, 40], [50, 60]])
Allocation_Factors = np.array([[0.5, 0.8], [1.2, 0.9]])

Resource_Distribution = np.dot(Resources_Matrix, Allocation_Factors)
print("A.1 Resource Distribution:\n", Resource_Distribution)

Shift_A_Production = np.array([100, 120, 130, 90, 110, 150, 140])
Shift_B_Production = np.array([80, 90, 100, 110, 95, 105, 120])

Total_Production = Shift_A_Production + Shift_B_Production
print("\nA.2 Total Production:\n", Total_Production)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

test_values = np.array([-10, 0, 10])
print("\nA.3 Sigmoid Output:", sigmoid(test_values))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

print("\nA.4 Sigmoid Gradient:", sigmoid_derivative(test_values))
```
